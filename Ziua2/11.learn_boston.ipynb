```python
y = df[["medv"]]
y
```

```python
x = df.drop("medv", axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
linear_model = LinearRegression()
linear_model
```

```python
linear_model.fit(x_train, y_train)
```

```python
linear_model.score(x_test, y_test)
```

```python
knn_model = KNeighborsRegressor()
```

```python
knn_model.fit(x_train, y_train)
```

```python
knn_model.score(x_test, y_test)
```

```python
y_prezis_knn= knn_model.predict(x_test)
y_prezis_knn  
```

```python
y_prezis_linear = linear_model.predict(x_test)
y_prezis_linear  # y - rezultat in urma folosirii modelului
```

```python
pd.DataFrame({"y_real": np.array(y_test).flatten(), "y_prezis_linear": np.array(y_prezis_linear).flatten()})
```
