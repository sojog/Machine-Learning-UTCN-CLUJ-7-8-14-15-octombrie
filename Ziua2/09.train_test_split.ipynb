```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.linspace(1, 100, num=25)
arr
```

```python
arr_train = arr[:20]
arr_train
```

```python
arr_test = arr[20:]
arr_test
```

```python
plt.scatter(arr_train, arr_train)
plt.scatter(arr_test, arr_test)
```

```python
from sklearn.model_selection import train_test_split
```

```python
arr_train, arr_test = train_test_split(arr)
plt.scatter(arr_train, arr_train)
plt.scatter(arr_test, arr_test)
```
