```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
df = pd.read_csv("boston_data.csv", index_col="ID")
df
```

```python
x = df.drop("medv", axis=1)
y = df[["medv"]]
```

```python
knn_model = KNeighborsRegressor(n_neighbors=7)
knn_model
```

```python
model_list = [ KNeighborsRegressor(n_neighbors=i) for i in range(1, 10) ]
model_list
```

```python
model_list[0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
for model in model_list:
    model.fit(x_train, y_train)
```

```python
model_list[3]
```

```python
for model in model_list:
    print("Scorul pentru modelul cu k=", model.n_neighbors, " este:",  model.score(x_test, y_test))
```
