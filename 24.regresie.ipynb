```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([10, 15, 20, 25, 30, 35])
y = np.array([10, 15, 20, 25, 30, 35])

plt.plot(x, y);
plt.scatter(x, y);
```

```python
x = np.array([10, 15, 20, 25, 30, 35])
y = np.array([10, 15, 20, 25, 30, 35])

plt.plot(x, y, color="red");
plt.scatter(x, y);
```

```python
x = np.array([10, 15, 20, 25, 30, 35])
y = np.array([20, 15, 30, 15, 30, 25])

plt.plot(x, y, color="red");
plt.scatter(x, y);
```
