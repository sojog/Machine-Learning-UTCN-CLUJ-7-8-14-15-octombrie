```python
from scipy import stats
```

```python
result = stats.linregress(suprafete, preturi)
slope = result.slope
intercept = result.intercept
```

```python
result
```

```python
plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi)

plt.plot(suprafete, suprafete * 568.6322001056895 + 53383.16892725031)
```

```python
plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi)

plt.plot(suprafete, suprafete * slope + intercept)
```

```python
plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi)
plt.xlabel("suprafete - INDEPENDENTE")
plt.ylabel("preturi - DEPENDENTE")

plt.plot(suprafete, suprafete * slope + intercept)
```

```python
def f(x):
    return x * slope + intercept
```
